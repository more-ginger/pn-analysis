```python
import json
import os

import pandas as pd

from supabase import create_client, Client
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
# Load variables from .env into environment
load_dotenv()
```

## Import JSON as string using Supabase API

```python
url: str = os.getenv("project_url")
key: str = os.getenv("service_key")
```

```python
supabase: Client = create_client(url, key)
```

```python
with open("./places/coverage_by_region.json", "r") as f:
    nested_data = json.load(f)
```

```python
nested_data_str = json.dumps(nested_data)
```

```python
data = {"data":nested_data_str}
```

```python
supabase.from_('regions').insert(data).execute()
```

## Import CSV files using SQLAlchemy

```python
user = os.getenv("user")
password = os.getenv("password")
host = os.getenv("host")
port = os.getenv("port")
dbname = os.getenv("dbname")
```

```python
coverage_by_country = pd.read_csv("./places/coverage_by_country.csv")
```

```python
coverage_by_country
```

```python
engine = create_engine(f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{dbname}")
```

```python
coverage_by_country.to_sql("coverage_by_country", engine, if_exists="replace", index=False)
```
