```python
import pandas as pd
import numpy as np
import ast

from collections import Counter
```

```python
nyt_coverage = pd.read_csv("../../input-data/temp-data.csv")
```

```python
nyt_coverage
```

```python
nyt_coverage["headline"] = nyt_coverage["headline"].apply(ast.literal_eval)
```

```python
nyt_coverage["keywords"] = nyt_coverage["keywords"].apply(ast.literal_eval)
```

```python
nyt_coverage["headline"] = nyt_coverage["headline"].apply(lambda x: x.get("main"))
```

```python
nyt_coverage = nyt_coverage[["clean_id", "headline", "web_url", "snippet", "lead_paragraph", "pub_date", "keywords", "image_url", "section_name", "word_count"]]
```

```python
nyt_coverage["keywords_type"] = nyt_coverage["keywords"].apply(lambda x: [kw.get("name") for kw in x])
    
```

```python
nyt_coverage["keywords"] = nyt_coverage["keywords"].apply(lambda x: [kw.get("value") for kw in x])
```

```python
nyt_coverage
```

```python
nyt_coverage = nyt_coverage[nyt_coverage["keywords"].apply(len) != 0]
```

```python
lists_of_structures = nyt_coverage["keywords_type"]
```

```python
# Step 1: Convert each list into a tuple (so it's hashable and can be counted)
structure_tuples = [tuple(lst) for lst in lists_of_structures]

# Step 2: Count the frequency of each structure
structure_counts = Counter(structure_tuples)

# Step 3: Convert to a pandas DataFrame
similar_patterns = pd.DataFrame(structure_counts.items(), columns=['Structure', 'Count'])

# Optional: Sort by most frequent
similar_patterns = similar_patterns.sort_values(by='Count', ascending=False).reset_index(drop=True)

print(similar_patterns)
```

```python
similar_patterns["Structure length"] = similar_patterns["Structure"].apply(lambda x: len(x))
```

```python
similar_patterns
```

```python
filtered_patterns = similar_patterns[similar_patterns['Structure'].apply(lambda x: 'glocations' in x)]
filtered_patterns
```

```python
# Get the most common structure
target_structure = similar_patterns.iloc[11102]['Structure']
```

```python
# Make sure target_structure is a tuple for comparison
target_structure = tuple(target_structure)

# Create a boolean mask and filter
matched_rows = nyt_coverage[nyt_coverage['keywords_type'].apply(lambda x: tuple(x) == target_structure)]
```
